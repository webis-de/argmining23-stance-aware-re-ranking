```python
from math import nan

from pandas import DataFrame, read_csv
from re import compile
```

```python
df_original: DataFrame = read_csv("data/metrics-reranking-all-original.csv")
df_original["run"] = df_original["name"].apply(lambda name: name.split(" + ")[0])
df_original["team"] = df_original["run"].apply(lambda run: run.split(" ")[0])
df_original
```

```python
df_gt: DataFrame = read_csv("data/metrics-reranking-all-ground-truth.csv")
df_gt["run"] = df_gt["name"].apply(lambda name: name.split(" + ")[0])
df_gt["team"] = df_gt["run"].apply(lambda run: run.split(" ")[0])
df_gt
```

```python
df_gpt: DataFrame = read_csv("data/metrics-reranking-all-gpt3.csv")
df_gpt["run"] = df_gpt["name"].apply(lambda name: name.split(" + ")[0])
df_gpt["team"] = df_gpt["run"].apply(lambda run: run.split(" ")[0])
df_gpt
```

```python
cols_combinations = [
    ["nDCG@5 rel."],
    ["nDCG@5 qual."],
    ["nDCG@5 rel.", "nDCG@5 qual."],
    ["alpha_nDCG@5 rel."],
    ["alpha_nDCG@5 qual."],
    ["alpha_nDCG@5 rel.", "alpha_nDCG@5 qual."],
]
```

```python
eval_data = []
```

```python
for labels, df_labels in {"original": df_original, "ground truth": df_gt, "GPT-3": df_gpt}.items():
    df_labels = df_labels[df_labels["name"].str.endswith(" + stance-first@5 + alternating-stance@5")]
    for cols in cols_combinations:
        df_tmp = df_labels.copy()
        for col in cols:
            df_tmp = df_tmp[df_tmp[f"{col} test"].notna()]
            if len(df_tmp) > 0:
                df_tmp = df_tmp[df_tmp[f"{col} test"].str.contains("1")]
        eval_data.append({
            "labels": labels,
            "cols": " & ".join(cols),
            "num_significant": len(df_tmp),
            "prop_significant": len(df_tmp) / len(df_labels["run"].unique()),
        })
```

```python
def target_f1(name: str) -> float:
    parts = name.split(" + ")
    for part in parts:
        if part.startswith("randomize"):
            match = compile(r"randomize\(F1<=(.*)\)").match(part)
            if not match:
                return 1.0
            return float(match.group(1))
    return nan
```

```python
df_f1: DataFrame = read_csv("data/metrics-reranking-all-f1.perquery.csv")
df_f1["run"] = df_f1["name"].apply(lambda name: name.split(" + ")[0])
df_f1["team"] = df_f1["run"].apply(lambda run: run.split(" ")[0])
df_f1["target_f1"] = df_f1["name"].apply(target_f1).fillna(1)
df_f1
```

```python
for cols in cols_combinations:
    for f1 in df_f1["target_f1"].unique():
        df_tmp = df_f1.copy()
        df_tmp = df_tmp[df_tmp["target_f1"] == f1]
        for col in cols:
            df_tmp = df_tmp[df_tmp[f"{col} test"].notna()]
            if len(df_tmp) > 0:
                df_tmp = df_tmp[df_tmp[f"{col} test"].str.contains("1")]

        df_tmp = df_tmp.groupby("run").first().reset_index()

        eval_data.append({
            "labels": f"randomized F1<={f1:.2f}",
            "cols": " & ".join(cols),
            "num_significant": len(df_tmp),
            "prop_significant": len(df_tmp) / len(df_f1["run"].unique()),
        })
```

```python
df_eval = DataFrame(eval_data)
df_eval.to_csv("data/num_significant.csv")
df_eval
```
